#Softmax activation is one of the most common activation functions used for classification problems. This week's question is focussed around the same.

Consider the following set of graphs.

The following set of graphs correspond to the following
- Softmax activation of logits
- Raw Logits/ Output scores
- Softmax activation of logits scaled down by a factor of 2
- Softmax activation of logits scaled up by a factor of 2

Now these are not in order, the question is to find out which graph corresponds to which scenario from the ones mentioned above from the options below

- A: Logits; C: Softmax; B: Softmax of scaled down logits; D: Softmax of scaled up logits
- B: Logits; A: Softmax; D: Softmax of scaled up logits; C: Softmax of scaled down logits
- C: Logits; B: Softmax; A: Softmax of scaled down logits; D: Softmax of scaled up logits
- None of the Above

```python
import pandas as pd
import numpy as np
import torch
np.set_printoptions(suppress = True)
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use("fivethirtyeight")
import warnings; warnings.filterwarnings("ignore")
```

```python
classes  = ["car", "bike", "plane", "dog", "cat", "frog", "apple", "mango", "grape", "bottle"]
```

```python
np.random.seed(10)
x = np.random.randn(10)
t = torch.Tensor(x)
t
```

```python
t.softmax(dim = -1)
```

```python
(t * 2).softmax(dim = -1)
```

```python
(t * .5).softmax(dim = -1)
```

```python
fig, ax = plt.subplots(2, 2, figsize = (10,7))
ax = ax.flat

ax[0].bar(classes, t); ax[0].set_xticklabels(classes, rotation = 45, fontsize = 8); ax[0].set_title("Option A", fontsize = 8)

ax[1].bar(classes, (t * 0.5).softmax(dim = -1)); ax[1].set_xticklabels(classes, rotation = 45, fontsize = 8); ax[1].set_title("Option B", fontsize = 8)

ax[2].bar(classes, t.softmax(dim = -1)); ax[2].set_xticklabels(classes, rotation = 45, fontsize = 8); ax[2].set_title("Option C", fontsize = 8)

ax[3].bar(classes, (t * 2).softmax(dim = -1)); ax[3].set_xticklabels(classes, rotation = 45, fontsize = 8); ax[3].set_title("Option D", fontsize = 8);
```
